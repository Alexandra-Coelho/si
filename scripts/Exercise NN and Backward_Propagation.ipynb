## NN model for the breast-bin.csv dataset

```python
path = 'C:/Users/ASUS/Desktop/Bioinfo/2ano/Sistemas Inteligentes/si/datasets/breast-bin.csv'
breast = read_csv(path, sep = ",", features = False, label = True)
print(breast.shape())
```

```python
from sklearn.preprocessing import StandardScaler
breast.X = StandardScaler().fit_transform(breast.X) 
train, test = train_test_split(breast, test_size=0.2, random_state=32)
```

```python
l1_breast = Dense(input_size=9, output_size=9)
l2_breast = Dense(input_size=9, output_size=4)
l3_breast = Dense(input_size=4, output_size=1)

l1_act = ReLUActivation()
l2_act = ReLUActivation()
l3_act = Sigmoid_Activation()

layers_breast = [l1_breast, l1_act, l2_breast, l2_act, l3_breast, l3_act]
```

```python
nn_model_bin = NN(layers_breast, loss_function = cross_entropy, loss_derivation = cross_entropy_derivative, verbose=True)  
```

```python
nn_model_bin.fit(train)
```

```python
pred = nn_model_bin.predict(test)
pred
```

```python
pred = np.where(pred > 0.5, 1, 0)
print('Score: ', round(accuracy(test.y , pred), 4))
```

## NN model for the cpu.csv dataset

```python
path = 'C:/Users/ASUS/Desktop/Bioinfo/2ano/Sistemas Inteligentes/si/datasets/cpu.csv'
cpu = read_csv(path, sep = ",", features = True, label = True)
print(cpu.shape())
```

```python
from sklearn.preprocessing import StandardScaler
cpu.X = StandardScaler().fit_transform(cpu.X)  # para normalizar dados
train, test = train_test_split(cpu, test_size=0.2, random_state=32)
```

```python
l1_cpu = Dense(input_size=6, output_size=6)
l2_cpu = Dense(input_size=6, output_size=4)
l3_cpu = Dense(input_size=4, output_size=1)

l1_activation = ReLUActivation()
l2_activation = ReLUActivation()
l3_activation = LinearActivation()
layers_cpu = [l1_cpu, l1_activation, l2_cpu, l2_activation, l3_cpu, l3_act]
```

```python
nn_model_reg = NN(layers_cpu, verbose = True) 
```

```python
nn_model_reg.fit(train)
```

```python
pred = nn_model_reg.predict(test)
```

```python
print('Score: ', round(rmse(test.y, pred), 4))
```
